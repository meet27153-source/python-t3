WAP to copy odd lines from 1 file to another

```python
f=open("hola.txt",'r')
l=f.read()
print(l)
l=l[::2]
print(l)
f.close()
```

WAP to replace a word in a file

```python
s=input("Enter old string:")
s1=input("Enter new string:")
f=open("hola.txt",'r')
data=f.read()
data=data.replace(s,s1)
f.close()
f1=open("hola.txt",'w')
f1.write(data)
f1.close()
```
