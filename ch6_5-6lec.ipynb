Write a pajor problem of your solution prompt for a file name and display the text file 5 lines at a time causing each time      asked user to press any key to continue and press q to quit

```python
file=open(input("Enter file name:"))
lines=5
for i,line in enumerate(file,1):
    print(i,":",line)
    if i%lines==0:
        x=input("Enter any key to continue or press q to quit:")
        if x=='q' or x=='Q':
            break
        else:
            continue
```

```python
with open("file1.txt") as f1:
    s1=f1.read()
with open("file2.txt") as f2:
    s2=f2.read()
character=1
lines=1
for i in range(len(s1)):
    if s1[i]==s2[i]:
        character+=1
        if s1[i]=='\n':
            lines+=1
            character=1
    else:
        print("Lines:",lines,"characters:",character)
        character+=1
```

```python
f=open("addition.txt")
d=open("output.txt",'w')
while True:
    b=f.readline()
    if len(b)!=0:
        if b[0]=='#':
            continue
        else:
            if '#' in b:
                for i in range(1,len(b)):
                    if b[i]=='#':
                        d.write(b[0:i]+"\n")
            else:
                d.write(b)
    else:
        break
f.close()
d.close()
```

WAP to search for a string in text file 

```python
f=open("output.txt")
s=input("Enter string to search:")
data=f.read()
i=data.find(s)
if i==-1:
    print("String not found")
else:
    print("String found at index:",i)
f.close()
```

WAP to extract all 4 letter words that starts and end with same letter from a given text fle

```python
f=open("ww.txt")
s1=f.read().split()
for i in s1:
    if (len(i)==4) and (i[0]==i[-1]):
        print(i)
f.close()
```
