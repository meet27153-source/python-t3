WAP to count the frequency of the words from given text file

```python
f=open("hello.txt")
l=f.read().split()
d={}
for i in l:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
        
for i in d:
    print(i,"-",d[i],end=",")
f.close()
```

WAP to accept string or lines from user if the user enters end,each string or line entered by user should be a new line in file.
Save all the lines in text file and display only those lines which begins with capital letters

```python
f=open("new.txt",'w+')
while True:
    s=input("Enter data(for quit enter END):")
    if s=='END':
        break
    else:
        f.write(s+'\n')
f.seek(0)
for i in f:
    if i[0].isupper():
        print(i[:-1])
f.close()
```

WAP to read text file and print only words starting with i in reverse order

```python
f=open("new.txt")
l=f.read().split()
new=""
for u in l:
    if u[0]=='i' or u[0]=='I':
        new+=u[::-1] + " "
    else:
        new+=u + " "
print(new)
f.close()
```
